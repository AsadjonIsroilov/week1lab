```python
scores = [72, 85, 90, 68, 88]
print("Scores:", scores)


average = sum(scores) / len(scores)
print("Average score:", average)


print("Highest score:", max(scores))
print("Lowest score:", min(scores))
```

```python
study_hours = [1, 2, 3, 4, 5]
exam_scores = [50, 55, 65, 70, 80]


print("\nStudy Hours and Exam Scores:")
for hours, score in zip(study_hours, exam_scores):
    print(hours, score)


print("\nInput → Output:")
for hours, score in zip(study_hours, exam_scores):
    print(f"Input: {hours} hours → Output: {score} marks")
```

```python
def pass_fail(score):
    if score >= 60:
        return "Pass"
    else:
        return "Fail"

scores = [45, 67, 82, 59]

for s in scores:
    print(f"Score: {s} → {pass_fail(s)}")
```

```python
sizes = [500, 800, 1000, 1200]
prices = [50, 80, 100, 130]

for size, price in zip(sizes, prices):
    print(f"Size: {size} sq ft → Price: {price}")


positive_trend = all(prices[i] > prices[i-1] for i in range(1, len(prices)))

if positive_trend:
    print("Positive trend")
else:
    print("No positive trend")
```

```python
study_hours = [2, 4, 6, 8]
scores = [55, 65, 75, 85]

increases = [scores[i] - scores[i-1] for i in range(1, len(scores))]
average_increase = sum(increases) / len(increases)
print("Average increase for every 2 hours:", average_increase)

predicted_score = scores[-1] + average_increase
print("Predicted score for 10 hours:", predicted_score)
```

```python
scores = [35, 58, 62, 74, 89, 45]

for score in scores:
    if score < 50:
        category = "Low"
    elif 50 <= score <= 70:
        category = "Medium"
    else:
        category = "High"

    print(f"Score: {score} → Category: {category}")
```

```python
ages = [18, 20, 22, 24, 26]
heights = [160, 165, 170, 175, 180]

for age, height in zip(ages, heights):
    print(f"Age: {age} → Height: {height} cm")

avg_age = sum(ages) / len(ages)
print("Average age:", avg_age)

avg_height = sum(heights) / len(heights)
print("Average height:", avg_height)

height_trend = all(heights[i] > heights[i-1] for i in range(1, len(heights)))

if height_trend:
    print("Height appears to increase with age.")
else:
    print("Height does not appear to increase with age.")
```
